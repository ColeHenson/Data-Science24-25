**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python
with pm.Model() as model_weight:

    #priors for the linear part of our model
    α = pm.Normal("α", mu=0, sigma=50)
    β = pm.Normal("β", mu=0, sigma=100)

    #this is the linear part of our model
    μ = pm.Deterministic("μ", α + β * howell.height)

    #prior for the standard deviation of our likelihood
    #Cauchy is a T dist with nu = 1
    σ = pm.HalfCauchy("σ", 10)

    #likelihood
    y_pred = pm.Normal("y_pred", mu=μ, sigma=σ, observed=howell.weight)

    #inference data object
    idata_weight = pm.sample()
```

```python
az.plot_posterior(idata_weight, var_names=["~μ"], figsize=(12, 3))
```

```python
#grabs 100 posterior samples
posterior = az.extract(idata_weight, num_samples=100)

# grabbing x values for graphing.
x_plot = xr.DataArray(
    np.linspace(howell.height.min(), howell.height.max(), 50),
    dims="plot_id"
    )

# this creates the expected line, the path we predict temperature and
#rented bike count deviates from
mean_line = posterior["α"].mean() + posterior["β"].mean() * x_plot

#These are a 50 lines that our model came up with
lines = posterior["α"] + posterior["β"] * x_plot

#set up plot
_, ax = plt.subplots()

#plots 50 other lines our model came up with.
ax.plot(x_plot, lines.T, c="C1", alpha=0.2, label="lines")

#plots the mean line
ax.plot(x_plot, mean_line, c="C0", label="mean line")

#plot the raw data
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)

#label axes and create legend
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

**Task6**:

Run a posterior predictive check for model_howell, plot the posterior predictive distribution, and comment on the fit of the model you made in Task5.

```python
# posterior predictive check
pm.sample_posterior_predictive(idata_weight, model=model_weight,  extend_inferencedata=True)
```

```python
az.plot_ppc(idata_weight, num_pp_samples=200, colors=["C1", "C0", "C1"])
```

```python
#copied from notes AK and changes var names

#plots the posterior predictive distribution

#grabs the expected line, or line of best fit formula
mean_line = idata_weight.posterior["μ"].mean(("chain", "draw"))

#creates some x values to run through our line formula
heights = np.random.normal(howell.height.values, 0.01)
idx = np.argsort(heights)
x = np.linspace(heights.min(), heights.max(), 15)

#grabs the 94% HDI and 50% HDI, and sets them up for graphing.
y_pred_q = idata_weight.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)
y_hat_bounds = iter(
    [
        PchipInterpolator(heights[idx], y_pred_q[i][idx])(x)
        for i in range(4)
    ]
)

#plots raw data and our line of best fit
_, ax = plt.subplots()
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.plot(howell.height[idx], mean_line[idx], c="C0")


#graphs the 94% and 50% HDIs
for lb, ub in zip(y_hat_bounds, y_hat_bounds):
    ax.fill_between(x, lb, ub, color="C1", alpha=0.5)

#labels
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

The model actually does a pretty good job of fitting the data in that the data is contained withing the standard deviations of the prediction for the most part. The model does not really account for the shape of the date and the increasing rate of change in the weight as result of increases in height. The data is not really linear to trying to model it linearly is not going to capture the non linear elements of the data. The posterior predictive check does not do very well as it predicts negative weights which impossible. The shape of the posterior predictive also completely misses the spike in low weights.

**Task7**:

Create and visualize the posterior of a generalized linear model that improves the on the fit of the model you made in Task5.

```python
sns.violinplot(howell.weight)
```

```python
with pm.Model() as model_sqr_weight:

    #priors for the linear part of our model
    α = pm.Normal("α", mu=0, sigma=10)
    β = pm.Normal("β", mu=0, sigma=10)

    #this is the linear part of our model
    μ = pm.Deterministic("μ", α + β * pm.math.sqr(howell.height))

    #prior for the standard deviation of our likelihood
    #Cauchy is a T dist with nu = 1
    σ = pm.HalfCauchy("σ", 10)

    #likelihood
    y_pred = pm.Normal("y_pred", mu=μ, sigma=σ, observed=howell.weight)

    #inference data object
    idata_sqr_weight = pm.sample()
```

```python
pm.sample_posterior_predictive(idata_sqr_weight, model=model_weight,  extend_inferencedata=True)
```

```python
az.plot_ppc(idata_sqr_weight, num_pp_samples=200, colors=["C1", "C0", "C1"])
```

```python
#copied from notes AK and changes var names

#plots the posterior predictive distribution

#grabs the expected line, or line of best fit formula
mean_line = idata_sqr_weight.posterior["μ"].mean(("chain", "draw"))

#creates some x values to run through our line formula
heights = np.random.normal(howell.height.values, 0.01)
idx = np.argsort(heights)
x = np.linspace(heights.min(), heights.max(), 15)

#grabs the 94% HDI and 50% HDI, and sets them up for graphing.
y_pred_q = idata_sqr_weight.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)
y_hat_bounds = iter(
    [
        PchipInterpolator(heights[idx], y_pred_q[i][idx])(x)
        for i in range(4)
    ]
)

#plots raw data and our line of best fit
_, ax = plt.subplots()
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.plot(howell.height[idx], mean_line[idx], c="C0")


#graphs the 94% and 50% HDIs
for lb, ub in zip(y_hat_bounds, y_hat_bounds):
    ax.fill_between(x, lb, ub, color="C1", alpha=0.5)

#labels
ax.set_xlabel("height")
ax.set_ylabel("weight")
```
